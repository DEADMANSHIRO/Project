```python
def y(a, b, x):
    return a * np.log(b * x)

a = np.arange(0.1, 5.0, 0.1)  # Начинаем с 0.1, чтобы избежать log(0)
b = 2  # Фиксируем b, иначе непонятно, какой брать
x = np.arange(0.1, 5.0, 0.1)

plt.figure()
plt.subplot(211)
plt.plot(x, y(a, b, x))  # x - аргумент, y - значение функции
plt.xlabel("Ось x")
plt.ylabel("Ось y")
plt.title("График функции y = a*ln(b*x)")
plt.grid()
plt.show()
```

```python
def f(a,phi):
    return a*phi

params = [(2),
          (8),
          (32) ]
phi = np.linspace(-10 * np.pi, 10 * np.pi, 10000) 

plt.figure(figsize=(6,6))

for i, (a) in enumerate(params,1):
    plt.polar(phi, f(a,phi), label=r"$r = a\phi$")  # Построение в полярных координатах

plt.legend()
plt.title("Спираль Архимеда")
plt.show()    
    


```

```python
def f(a,b,phi):
    return  a*np.cos(phi)+b

phi = np.linspace(-10 * np.pi, 10 * np.pi, 1000)

parametrs = [(2,4),
            (1,1.5),
            (6,1)]
                
plt.figure()

for i, (a,b) in enumerate(parametrs,1):
    plt.polar(phi, f(a,b,phi), label=fr"$r = {a}\cos(\phi) + {b}$")    

plt.legend()
plt.show()
```

```python
#Polar Rose for Koziryanus 

def polar_rose(a, phi):
    return np.sin(a * phi)

# Задаём параметры (a, phi_max)
params = [
    (6, 2 * np.pi),
    (7/4, 8 * np.pi),
    (3/4, 8 * np.pi)
]

plt.figure(figsize=(6, 6))

for a, phi_max in params:
    phi = np.linspace(0, phi_max, 500)  
    r = polar_rose(a, phi)    
    plt.polar(phi, r, label=fr"$r = \sin({a} \cdot \phi)$")

plt.legend()
plt.show()
```

```python
def f(r, t):
    x = r * np.cos(t)
    y = r * np.sin(t)
    return x, y

t = np.linspace(0, 2 * np.pi, 500)  
r = 5  # Фиксированное значение r (можно сделать массивом, если нужно)

x, y = f(r, t)  # Вычисляем координаты

plt.figure()
plt.plot(x, y)  # Теперь передаём x и y отдельно
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Окружность радиуса r")
plt.axis("equal")  # Сохраняем пропорции
plt.show()


```

```python
def f(a,b,c,d):
    u = np.linspace(-2*np.pi,2*np.pi, 100)
    v = np.linspace(-np.pi,np.pi, 50)
    U, V = np.meshgrid(u, v)

    x = np.cos(U)*np.cos(V) + a*np.cos(U)*(b+ np.sin(b*U)/c)
    y = np.sin(U)*np.cos(V) + a*np.sin(U)*(b+ np.sin(b*U)/c)
    z = np.sin(V) + d*np.cos(b*U)
    return x,y,z

params = [(3,1.5,2,2),
         (3,1.5,1,2),
         (4,5,6,1),
         (1,2,3,4)]


fig = plt.figure(figsize = (15,15))
for i, (a,b,c,d) in enumerate(params,1):
    ax = fig.add_subplot(1, 4, i, projection='3d')
    x,y,z = f(a,b,c,d)
    ax.plot_surface(x, y, z, cmap='viridis')  
    ax.set_title(f"a={a}, b={b}, c={c}, d={d}")
    ax.view_init(elev=30, azim=60)


plt.show()
```

```python
# Параметр t от 0 до 2π
t = np.linspace(0, 2 * np.pi, 500)

# Параметрические уравнения трилистника
x = np.sin(t) + 2 * np.sin(2*t)
y = np.cos(t) - 2 * np.cos(2*t)
z = -np.sin(3*t)

# Создаём 3D-график
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Рисуем трилистник
ax.plot(x, y, z, color='blue', linewidth=2)

# Настройки осей
ax.set_title("Трилистник")
ax.view_init(elev=30, azim=30)  # Угол просмотра

plt.show()
```

```python
def f(a):
    u = np.linspace(0, 2*np.pi, 100)
    v = np.linspace(-1, 1, 100)
    U,V = np.meshgrid(u,v)

    x = (1 + (V/a)*np.cos(U/a))*np.cos(U)
    y = (1 + (V/a)*np.cos(U/a))*np.sin(U)
    z = (V/a)*np.sin(U/a)
    
    return x,y,z

params = [(2),
          (3),
          (4),
          (5)]

fig = plt.figure(figsize = (15,15))
for i, a in enumerate(params,1):
    ax = fig.add_subplot(1, 4, i, projection='3d')
    x,y,z = f(a)
    ax.plot_surface(x, y, z, cmap='viridis')  
    ax.set_title(f"a={a}")
    ax.view_init(elev=30, azim=60)

plt.show()
    
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Задаем параметры
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, 2 * np.pi, 100)
u, v = np.meshgrid(u, v)

# Радиус
r = 2 

# Функции для трилистника
x = (2 + r * np.cos(v)) * np.cos(3 * u)
y = (2 + r * np.cos(v)) * np.sin(3 * u)
z = r * np.sin(v)

# Построение графика
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

# Оформление
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Параметрическая поверхность трилистника")

plt.show()
```
